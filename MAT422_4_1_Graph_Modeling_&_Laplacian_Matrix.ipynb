# Graph Models

Graph models are incredibly useful tools to analyse networks and to build networks to successfully operate and communicate. These are commonly used for social media websites and other related applications in which networks are the important and underlying function for the system.

image.png

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.erdos_renyi_graph(n=30, p=0.5)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', font_size=8, font_color='black', font_weight='bold', edge_color='gray', linewidths=0.5)
plt.title("Random Graph for Network Analysis")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_nodes_from([1, 2, 3, 4])
nx.add_cycle(G, [1, 2, 3, 4])
G.add_edges_from([(5, 3), (8, 5), (2, 4), (6, 7)])
nx.draw(G)
plt.show()

G = nx.complete_graph(5)
nx.draw(G)
plt.show()
```

# Laplacian Matrix

A Laplacian matrix is also an incredibly important matrix in that can be effectively used within networks. It can be used to accurately analyse networks as well as setting networks up and allowing them to function efficiently. They are typically commonly used for creating and managing networks and determining how to upkeep them. These networks can be directed to communicate directly or indirectly through variosus methods which are done by laplacian matrices.

image.png

image.png

```python
H = nx.Graph()
H.add_edges_from([(10, 20), (20, 30), (30, 10), (30, 40), (40, 50), (50, 30)])

lap_matrix = nx.laplacian_matrix(H).toarray()

print("Modified Laplacian Matrix:")
print(lap_matrix)

layout = nx.spring_layout(H)
nx.draw(H, layout, with_labels=True, node_size=800, node_color='lightgreen', font_size=10, font_color='darkblue', font_weight='normal', edge_color='silver', linewidths=1.0)
plt.title("Graph with Modified Laplacian Matrix")
plt.show()
```
