# Low-Rank Matrix Approximations

Observing the distance between two matrices, the approximation of low-rank matrices can be done. This is done by using the processes and methods from the Singular Value Decomposition and keeping k to the left vectors of U, diagonal values of S, and the top vectors of V.

image.png

```python
import numpy as np

A = np.random.rand(10, 4)

U, Sigma, VT = np.linalg.svd(A)

rank = 2

LRA = np.dot(U[:, :rank], np.dot(np.diag(Sigma[:rank]), VT[:rank, :]))

print("Original Matrix:")
print(A)
print("Low-rank Approximation:")
print(LRA)
```

# Covariance Matrix

In order to successfully carry out a principal component analysis, accurately setting up the pxN matrix of observation as well as finding the covariance matrix, which is the pxp matrix S, then the process of PCA can be accurately conducted.

```python
import numpy as np

X = np.random.rand(10, 4)

# Calculate the covariance matrix
cov_matrix = np.cov(X)

print("X:")
print(X)
print("Covariance Matrix:")
print(cov_matrix)
```

# Total Variance

Once determining the columns of the pxN data matrix and obtaining the mean-deviation, the covariance matrix can be st  by S, from  which the calculated values of t and r can be inputted into the equation tr(S)in order to find total variance, which is the sum of variances on the diagonal of S.

image.png

```python
import numpy as np

X = np.random.rand(10, 4)
meanX = np.mean(X)
stdX = np.std(X)
Xsub = (X - meanX)
XX = Xsub / stdX

cov_matrix = np.cov(XX.T)

eigenvalues, _ = np.linalg.eig(cov_matrix)

variance = np.sum(eigenvalues)

print("Total Variance:")

print(variance)
```
