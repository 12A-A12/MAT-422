# QR Decomposition
The process of QR decomposition is a commonly used process that can be used to solve linear least squares problems. Within this process the Gram-Schmidt algorithim can be used to determine the orthonormal base span and as a result the QR decomposition can be obtained.

image.png

```python
import numpy as np
# Gram Schmidt Orthogonlization
u1 = np.array([1, 2, 3])
u2 = np.array([0, 5, 5])
u3 = np.array([0, 0, 5])
U = ([u1,u2,u3])
U2 = np.array(U)

np.linalg.norm(U)

product = U2/np.linalg.norm(U)

print("Vector is", product)
print("The vector U is not orthonormal")

# QR Decomposition
A = np.array(U)

print("Matrix A")
print(A)

v1 = np.array([0, 5, 5])
v2 = np.array([8, 5, 1])
v3 = np.array([2, 3, 3])

q, r = np.linalg.qr(A)

print('Q:', q)
print('R:', r)

# Verification of value

QR = np.dot(q, r)
print('QR:', QR)

r = q.T @ A
print('R is ', r)

QR = q @ r
print('QR is ', QR)
```

Above shows an example of QR decomposition, with the qr function being used to decompose matrix A in order to find the Q and R matrices, after which they can be multiplied in order to verify the accuracy of the code and solution, which it is found that QR is equal to A, thus indicating that A = QR per the equation given. This can further be vaarified using Gram-Schmidt where decomposition converts A into the product of an orthgonal matrix Q.
