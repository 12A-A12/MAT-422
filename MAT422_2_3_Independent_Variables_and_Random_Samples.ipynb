# Independent Random Variables
The values of two variables X and Y can be considered ot be independent random variables, which describes instances in which one of the variables occcuring or being true does not affect the probability of the other variable, thus introducting this concept of independence.

image.png


```python
import numpy as np

X = np.array([1, 2, 3])
probabilities_X = np.ones(3) / 3

Y = np.array(['A', 'B', 'C'])
probabilities_Y = np.array([0.4, 0.3, 0.3])

randomX = np.random.choice(X, p=probabilities_X)
randomY = np.random.choice(Y, p=probabilities_Y)

print(f"Random value for X: {randomX}")
print(f"Random value for Y: {randomY}")
```
