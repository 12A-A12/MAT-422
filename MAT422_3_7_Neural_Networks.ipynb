# Activation Functions

Using activation functions in neural networks, nodes can be used to create outputs within a given node using inputs or classifications, in functions called activation functions. This can be used to begin processes in the nodes and  neural network.

image.png

```python
x = np.linspace(-10, 10)

def binary(x):
    return np.heaviside(x, 1)

plt.plot(x, binary(x))
plt.title('Activation Function')
plt.show()
```

# Cost Function

Along with many other functions that can be used, another essential and integral function that can be used can be utilized by incorporating least squares which could be used to classify and calculate various equations to be incorporated into the Neural Network.

image.png

```python
def meansquare(true, pred):
    return np.mean((true - pred) ** 2)

true = np.array([2, 4, 6, 8, 10])
pred = np.array([3, 5, 7, 9, 11])

meanerror = meansquare(true, pred)
print(f"MSE: {meanerror}")
```
