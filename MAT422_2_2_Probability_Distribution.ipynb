# Discrete Random Variables

When dealing with random variables, they can be defined as a measurable function that can be definied and measured with probability. It can be described as a variable whose value is dependant on various outcomes of various occurences.

```python
outcomes = {(num1, num2): int(num1) + int(num2) for num1, num2 in product(map(str, range(1, 7)), repeat=2)}
occurrences = {total: 0 for total in range(2, 13)}
for total in outcomes.values():
    occurrences[total] += 1

total_simulated_outcomes = len(outcomes)
probabilities_df = pd.DataFrame([[value, occurrence / total_simulated_outcomes] for value, occurrence in occurrences.items()], columns=['Sum of Numbers', 'Probability'])

ax = sns.barplot(x='Sum of Numbers', y='Probability', data=probabilities_df)
```

The above is a simulated example of Discrete Random variables, with a distribution of numbers and their given probabilities for their respective categories.
